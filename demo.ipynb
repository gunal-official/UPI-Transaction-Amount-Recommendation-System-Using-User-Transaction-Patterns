```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('upi_transactions_dataset.csv')
user_profiles = pd.read_csv('user_behavior_profiles.csv')

# Basic analysis
print("Dataset Overview:")
print(f"Transactions: {len(df)}")
print(f"Users: {df['user_id'].nunique()}")
print(f"Categories: {df['category'].nunique()}")

# Visualizations
plt.figure(figsize=(12, 8))

# Transaction amounts by category
plt.subplot(2, 2, 1)
df.groupby('category')['amount'].mean().plot(kind='bar')
plt.title('Average Amount by Category')
plt.xticks(rotation=45)

# User transaction counts
plt.subplot(2, 2, 2)
user_counts = df['user_id'].value_counts()
plt.hist(user_counts, bins=20)
plt.title('Distribution of User Transaction Counts')

plt.tight_layout()
plt.show()
```
